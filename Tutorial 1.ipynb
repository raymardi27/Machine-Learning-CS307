```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(40)
```

```python
def gaussian(x, mu=0, sigma=1):
    constant = 1/(((2 * np.pi)** 0.5)*sigma)
    x = ((x - mu)/sigma) ** 2
    y = np.exp(-0.5 * x) * constant
    return y
```

```python
x = np.random.normal(5, 1, 10)
x.sort()
```

```python
likelihood = []; mu = [];
for i in range(0, 110):
    y = gaussian(np.array(x), i/10)
    prod = np.prod(y)
    likelihood.append(prod)
    mu.append(i/10)
    
plt.plot(mu, likelihood)
plt.title("Maximum likelihood estimation")
plt.xlabel("Mean")
plt.ylabel("Likelihood function")
plt.savefig("Max_likelihood_estimation")
```

```python
log_likelihood = np.log(np.array(likelihood))
plt.plot(mu, log_likelihood)
plt.title("Maximum likelihood estimation")
plt.xlabel("Mean")
plt.ylabel("Log likelihood function")
plt.savefig("Log_max_likelihood_estimation")
```
