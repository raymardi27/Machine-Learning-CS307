```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = [1, 2, 3]
y = [1.2, 1.9, 3.2]
plt.scatter(x, y)
plt.xlabel("X data points")
plt.ylabel("Y data points")
plt.title("Scatter plot")
plt.savefig("scatter")
```

```python
def get_max_likelihood_line(x, y):
    w1 = (len(x) * x.dot(y.T) - (np.sum(y) * np.sum(x)))/(len(x) * x.dot(x.T) - np.sum(x) ** 2)
    w0 = np.mean(y) - w1 * np.mean(x)
    return w1, w0
```

```python
x = np.array(x)
y = np.array(y)
```

```python
w1, w0 = get_max_likelihood_line(x, y)
```

```python
w1, w0
```

```python
plt.plot(x, w1 * x + w0)
plt.scatter(x, y, color="r")
plt.xlabel("X data points")
plt.ylabel("Y data points")
plt.title("Best fit line")
plt.savefig("best_fit_line")
```
